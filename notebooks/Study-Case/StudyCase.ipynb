```python
series_resto_status_summary = df_trx.groupby(['resto_id','status']).food_price.sum()
series_resto_order_cnt = df_trx.groupby(['resto_id']).size()
```

```python
type(series_resto_status_summary)
```

```python
df_resto_status = series_resto_status_summary.unstack('status')
```

```python
series_resto_order_cnt
```

```python
df_resto_status['order_cnt'] = series_resto_order_cnt
```

```python
df_resto_status2 = df_resto_status[df_resto_status['order_cnt']>3]
```

```python
df_resto_status2['total_value'] = df_resto_status2[['Cancelled','Completed']].sum(axis = 1)
```

```python
df_resto_status2['cancelled_pct'] = (df_resto_status2['Cancelled']/df_resto_status2['total_value']).fillna(0)
```

```python
df_resto_status2.sort_values('cancelled_pct', ascending = False)[:10]
```
