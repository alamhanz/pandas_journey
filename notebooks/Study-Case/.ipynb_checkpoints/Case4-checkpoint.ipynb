RFM (clustering) --> Recency, Frequency, Monetary

```python
## assume today is 2018-10-10
```

```python
df_food = pd.read_csv(PATH_DATA+'food_delivery_datasets.csv', parse_dates = ['date_time'])
```

```python
df_food.head()
```

```python
df_food.date_time.max()
```

```python
df_food['range_to_today'] = (pd.to_datetime('2018-10-10')-df_food['date_time']).dt.days
```

```python
df_food.head()
```

```python
measure_rfm = {'order_id':np.size,'food_price': np.sum, 'range_to_today':np.min}
rename_cols = {'order_id':'frequency','food_price': 'monetary', 'range_to_today':'recency'}
```

```python
df_rfm = df_food.groupby(['cust_id']).agg(measure_rfm)
df_rfm.rename(columns = rename_cols, inplace= True)
```

```python
df_rfm.sample(10)
```

```python
df_rfm.describe()
```

```python
df_rfm['ff'] = pd.cut(df_rfm['frequency'], bins = [0, 4, 6, 10, 100], labels= ['f1','f2','f3','f4'])
df_rfm['mm'] = pd.cut(df_rfm['monetary'], bins = [0, 2.221500e+05, 3.926000e+05, 7.511000e+05, 4.136200e+07], labels= ['m1','m2','m3','m4'])
df_rfm['rr'] = pd.cut(df_rfm['recency'], bins = [0, 13, 17, 24, 200], labels= ['r4','r3','r2','r1'])
```

```python
matrix_rfm = df_rfm.groupby(['rr','ff','mm']).size().unstack()
```

```python
matrix_rfm
```

```python
plt.figure(figsize = (13,10))
plt.title('HeatMap')
fig_save=sns.heatmap(data = matrix_rfm,
           annot = True)
```

```python
df_rfm[(df_rfm['ff'] == 'f4') & (df_rfm['mm'] == 'm4') & (df_rfm['rr'] == 'r3')].index
```

Cohort Analysis

```python
df_food.head()
```

```python
df_food['week_order'] = df_food['date_time'].dt.strftime('%W')
```

```python
## Find first week order of every customer 
df_customer_first_week = df_food.groupby('cust_id')['week_order'].min().reset_index()
df_customer_first_week.columns = ['cust_id','first_week_order']
```

```python
df_customer_first_week.sample(10)
```

```python
df_food.head()
```

```python
df_food2 = pd.merge(df_food,df_customer_first_week,on = 'cust_id', how = 'left')
```

```python
df_food2.head()
```

```python
df_food2['next_x_week'] = df_food2['week_order'].astype(int) - df_food2['first_week_order'].astype(int)
df_food2.head()
```

```python
df_cohort1 = df_food2.groupby(['first_week_order','next_x_week']).agg({'cust_id':pd.Series.nunique}).unstack()
```

```python
df_cohort1
```

```python
## in percent
df_cohort1_transpose = df_cohort1.transpose()
df_cohort2 = ((df_cohort1_transpose/df_cohort1_transpose.max())).transpose()
```

```python
df_cohort2
```

```python
df_cohort2.mean()
```

```python
plt.figure(figsize = (25,10))
plt.title('Cohort')
fig_save=sns.heatmap(data = df_cohort2,
           annot = True, fmt = '.2%')
```
