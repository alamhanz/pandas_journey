```python
df_csv[['order_id','delivery_fee']]
```

```python
Mu = df_csv['delivery_fee'].mean() ## --> average
Su = df_csv['delivery_fee'].std() ## --> Standard Deviation

df_csv['delivery_fee_SS'] = (df_csv['delivery_fee']-Mu)/Su ## --> calculate and assign to a new column
```

```python
df_csv[['order_id','delivery_fee','delivery_fee_SS']].head()
```

```python
round(df_csv['delivery_fee_SS'].mean(),3),round(df_csv['delivery_fee_SS'].std(),3)
```
