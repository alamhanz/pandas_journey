# 8. Aggregation - Pivot

```python
import pandas as pd
import numpy as np
```

```python
PATH_DATA = '../mydata/'
```

## Get data

```python
df_csv = pd.read_csv(PATH_DATA+'food_delivery_datasets.csv')
```

```python
df_csv.head()
# df_csv.head(10)
```

```python
df_csv0 = df_csv.copy()
```

## Aggregation 1

```python
group_cust = df_csv.groupby('cust_id')
```

```python
group_cust['order_id'].size()
```

```python
group_cust['food_price'].mean()
```

```python
group_cust['eta_seconds'].mean()
```

```python
## Notes
type(group_cust['eta_seconds'].mean())
```

```python
group_cust['eta_seconds'].mean().reset_index()
```

```python
group_cust[['food_price','eta_seconds']].mean()
```

```python
group_cust.agg({'food_price':np.std,'eta_seconds':np.mean})
```

## Aggregation 2

```python
group_typ2 = df_csv.groupby(['cust_id','resto_id'])
```

```python
group_typ2.agg({'food_price':np.std,'eta_seconds':np.mean})
```

```python
group_typ2.agg({'food_price':[np.std,np.size],'eta_seconds':np.mean})
```

```python
group_typ2.agg({'food_price':[np.std,np.size],'eta_seconds':np.mean}).reset_index()
```

## Pivot

```python
df_csv.head()
```

```python
# df_csv.pivot(index='cust_id', 
#                      columns='rating', 
#                      values='eta_seconds')
```

```python
group_typ3 = df_csv.groupby(['cust_id','rating'])
df_grp3 = group_typ3[['eta_seconds']].mean()
df_grp3.reset_index()
```

```python
df_grp3.reset_index().pivot(index='cust_id', 
                     columns='rating', 
                     values='eta_seconds')
```

```python
## other way
df_grp3.unstack()
```

```python
group_typ2.agg({'eta_seconds':np.mean}).unstack()
```
